```python
import math

free, win, offer = 1, 0, 1
w = {
    "n1": [0.5, -0.2, 0.3],
    "n2": [0.4, 0.1, -0.5],
    "out": [0.7, 0.2]
}

relu = lambda x: max(0, x)
sigmoid = lambda x: 1 / (1 + math.exp(-x))
n1 = relu(free*w["n1"][0] + win*w["n1"][1] + offer*w["n1"][2])
n2 = relu(free*w["n2"][0] + win*w["n2"][1] + offer*w["n2"][2])

# output calculation
output_raw = n1*w["out"][0] + n2*w["out"][1]
spam_probability = sigmoid(output_raw)

# prints with explanation
print("Neuron 1 output (n1):", n1)
print("Neuron 2 output (n2):", n2)
print("Raw output before sigmoid:", output_raw)
print("Final spam probability:", spam_probability)
```
